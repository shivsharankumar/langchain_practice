```python
import random
class NakliLLM:
    def __init__(self):
        # self.name = name
        print('LLM Created')

    def predict(self, prompt):
        response_list=['Delhi is the capital of India.','The sun rises in the east.','Python is a popular programming language.']
        return {'response': random.choice(response_list)}
```

```python
class NakliPromptTemplate:
    def __init__(self,template,input_variables):
        self.template = template
        self.input_variables = input_variables
        print('Prompt Template Created')

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template=NakliPromptTemplate(template='Write a {length} sentence about {country}?',input_variables=['length', 'country'])
formatted_prompt=template.format({'length': 'short', 'country': 'India'})
print(formatted_prompt)
```

```python
llm=NakliLLM()
response=llm.predict(formatted_prompt)
print(response)
```

```python
#nakli llm chain
class NakliLLMChain:
    def __init__(self, llm, prompt_template):
        self.llm = llm
        self.prompt_template = prompt_template
        print('LLM Chain Created')

    def run(self, input_dict):
        formatted_prompt = self.prompt_template.format(input_dict)
        response = self.llm.predict(formatted_prompt)
        return response['response']
```

```python
template=NakliPromptTemplate(template='Write a {length} sentence about {country}?',input_variables=['length', 'country'])
formatted_prompt=template.format({'length': 'short', 'country': 'India'})
print(formatted_prompt)
```

```python
llm=NakliLLM()
chain=NakliLLMChain(llm=llm, prompt_template=template)
response=chain.run({'length': 'short', 'country': 'India'})
print(response)
```
